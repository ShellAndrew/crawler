```python

graph = {
    'A': ['B', 'C'],
    'B': ['A', 'D'],
    'C': ['A', 'E'],
    'D': ['B', 'E'],
    'E': ['C', 'D']
}

def generate_path(graph):
    goal = 'E'
    start = 'A'
    frontier = []
    frontier.append(start)
    came_from = {}

    while frontier:
        current = frontier[0]
        
        if current == goal:
            break
        
        frontier.pop(0)
        
        for next in graph[current]:
            if next not in came_from:
                frontier.append(next)
                came_from[next] = current

    current = goal
    path = []
    while current != start:
        path.append(current)
        current = came_from[current]
    path.append(start)
    path.reverse()

    print(path)

generate_path(graph)
```
